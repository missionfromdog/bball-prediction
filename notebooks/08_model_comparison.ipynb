# Model Comparison - NBA Game Prediction

This notebook compares 7+ different classification models to find the best performer for NBA game prediction.

**Models Tested:**
1. XGBoost
2. LightGBM
3. HistGradientBoosting (sklearn)
4. Random Forest
5. Extra Trees
6. Logistic Regression
7. Ridge Classifier

**Evaluation:**
- Primary Metric: AUC-ROC
- Secondary: Accuracy, F1, Precision, Recall
- Probability Quality: Brier Score, Log Loss
- Performance: Training time, Prediction time
- Validation: 5-Fold Stratified Cross-Validation + Hold-out Test Set
